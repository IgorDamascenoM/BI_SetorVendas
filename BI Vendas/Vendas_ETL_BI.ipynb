```python
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import joblib 
import seaborn as sns
import sklearn as sk
```

```python
df_vendas = pd.read_excel('Vendas.xlsx')
```

```python
#Split da coluna, localidade, criando as novas colunas Pais e Continente
#Usando o Delimitador ( - )

df_vendas[['Pais','Continente']] = df_vendas['Localidade'].str.split(' - ', expand=True)
```

```python
#Removendo a antiga Localidade e uma coluna que esta vazia

df_vendas = df_vendas.drop(columns=['Localidade','Unnamed: 9'])
```

```python
#Removendo as linhas que tenha qualquer valor NaN

df_vendas.dropna(axis=0, how='any')
```
